### Floats versus integers

```python
x=torch.Tensor([1.2 , 2.5])
print(x)
print(x.type())
```

```python
y=torch.LongTensor([5,6])
print(y)
print(y.type())
```

```python
y=y.float()
print(y)
print(y.type())
```

```python
x=x.long()
print(x)
print(x.type())
```

### Reshaping a tensor

```python
x=torch.arange(10)
print(x)
```

```python
print( x.view(2,5) )
```

```python
print( x.view(5,2) )
```

### Note that the original tensor x was NOT modified

```python
print(x)
```

### To make the change permanent you need to create a new tensor

```python
y=x.view(5,2)

print(x)
print('')
print(y)
```

### Let check the sizes after slicing
